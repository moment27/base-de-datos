```python
import folium

lugares = {
    "Jockey Plaza": [-12.1025, -76.9681],
    "Salamanca": [-12.0700, -77.0050],
    "Parque de los Anillos": [-12.0500, -77.0300],
    "Aeropuerto": [-12.0219, -77.1143],
}

mapa = folium.Map(location=lugares["Jockey Plaza"], zoom_start=12)

for lugar, coord in lugares.items():
    folium.Marker(
        location=coord,
        tooltip=lugar,
        icon=folium.Icon(color="blue")
    ).add_to(mapa)

ruta = [
    lugares["Jockey Plaza"],
    lugares["Salamanca"],
    lugares["Parque de los Anillos"],
    lugares["Aeropuerto"]
]

folium.PolyLine(locations=ruta, color="blue", weight=5, opacity=0.7).add_to(mapa)

mapa
```
