```python
class NodoLugar:
    def __init__(self, nombre, categoria, coordenadas):
        self.nombre = nombre
        self.categoria = categoria
        self.coordenadas = coordenadas
        self.siguiente = None

class NodoDistrito:
    def __init__(self, nombre):
        self.nombre = nombre
        self.lugares = None
        self.siguiente = None

    def agregar_lugar(self, nombre, categoria, coordenadas):
        nuevo = NodoLugar(nombre, categoria, coordenadas)
        if not self.lugares:
            self.lugares = nuevo
        else:
            actual = self.lugares
            while actual.siguiente:
                actual = actual.siguiente
            actual.siguiente = nuevo

class ListaDistritos:
    def __init__(self):
        self.inicio = None

    def agregar_distrito(self, nombre):
        nuevo = NodoDistrito(nombre)
        if not self.inicio:
            self.inicio = nuevo
        else:
            actual = self.inicio
            while actual.siguiente:
                actual = actual.siguiente
            actual.siguiente = nuevo
        return nuevo  # Para luego añadir lugares

    def obtener_distrito(self, nombre):
        actual = self.inicio
        while actual:
            if actual.nombre == nombre:
                return actual
            actual = actual.siguiente
        return None
```

```python
import folium
from folium.plugins import MarkerCluster
import geocoder
from collections import defaultdict

def parse_coordenadas(coord):
    if isinstance(coord, (tuple, list)) and len(coord) == 2 and all(isinstance(x, (float, int)) for x in coord):
        return coord
    if isinstance(coord, str):
        try:
            lat, lon = map(float, coord.split(','))
            return (lat, lon)
        except Exception:
            pass
    raise ValueError(f"Coordenadas no válidas: {coord}")

def mostrar_mapa_distritos(distritos: ListaDistritos, centro_mapa):
    mapa = folium.Map(location=centro_mapa, zoom_start=13, tiles="CartoDB positron")

    #categorías
    categorias_grupos = defaultdict(lambda: folium.FeatureGroup(name="", show=False))

    actual_distrito = distritos.inicio

    while actual_distrito:
        lugar = actual_distrito.lugares
        primer_coords = None

        while lugar:
            try:
                coords = parse_coordenadas(lugar.coordenadas)
                if primer_coords is None:
                    primer_coords = coords

                grupo_categoria = categorias_grupos[lugar.categoria]
                grupo_categoria.layer_name = lugar.categoria  # Nombre visible en el control

                folium.Marker(
                    location=coords,
                    popup=f"{lugar.nombre} ({lugar.categoria}) - {actual_distrito.nombre}",
                    icon=folium.Icon(color="blue", icon="info-sign")
                ).add_to(grupo_categoria)

            except Exception as e:
                print(f"Error en {lugar.nombre}: {e}")
            lugar = lugar.siguiente

        # Banderita del distrito
        if primer_coords:
            folium.Marker(
                location=primer_coords,
                popup=f"<b>{actual_distrito.nombre}</b>",
                tooltip=actual_distrito.nombre,
                icon=folium.Icon(color="green", icon="flag")
            ).add_to(mapa)

        actual_distrito = actual_distrito.siguiente

    # Agrega cada grupo de categoría al mapa
    for grupo in categorias_grupos.values():
        mapa.add_child(grupo)

    # Ubicación actual
    ubicacion_actual = geocoder.ip('me').latlng
    if ubicacion_actual:
        folium.Marker(
            location=ubicacion_actual,
            popup="Tu ubicación actual",
            icon=folium.Icon(color="red", icon="glyphicon-user")
        ).add_to(mapa)

    # Control de capas
    folium.LayerControl(collapsed=False).add_to(mapa)

    # Script para mostrar/ocultar categorías según zoom
    mapa.get_root().html.add_child(folium.Element("""
    <script>
        var map = document.querySelectorAll('.leaflet-container')[0]._leaflet_map;
        map.on('zoomend', function() {
            var zoom = map.getZoom();
            var shouldShow = zoom >= 15;
            map.eachLayer(function(layer) {
                if (layer.options && layer.options.pane === "markerPane" && layer.getTooltip) {
                    if (shouldShow) {
                        if (!map.hasLayer(layer)) map.addLayer(layer);
                    } else {
                        if (map.hasLayer(layer)) map.removeLayer(layer);
                    }
                }
            });
        });
    </script>
    """))

    return mapa

# Centro del mapa
centro_mapa_general = parse_coordenadas(StaAnita[0]["coordenadas"])
mapa_general = mostrar_mapa_distritos(distritos, centro_mapa_general)
mapa_general
```
